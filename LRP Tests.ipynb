### Building our own shallow DNN using lower level Tensorflow

```python
n_inputs = 28*28  # MNIST
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

Since the data we shove into the neural net will change for every batch, we cannot create the data nodes just yet, so we use a placeholder. We add names so that the graphs in Tensorboard make sense.

Note the ``None`` in the shape, this allows arbitrary number of samples. So we wont have to update everytime we change the batch size

```python
reset_graph()

X = tf.placeholder(tf.float32, shape=(None,n_inputs), name = "X")
y = tf.placeholder(tf.int32, shape=(None), name="y")

# A scope is just a holding cell for a Tensorboard graph,
# nothing special
with tf.name_scope("dnn"):
    hidden1 = tf.layers.dense(X, n_hidden1,name="hidden1",
                              activation=tf.nn.elu)
    logits = tf.layers.dense(hidden1 , n_outputs, name = "outputs" )
```

Using cross entropy is traditional for multiclass classigfication, though theoretically any loss could be used. We could try a simple MSE later (although that will require tweaking the output layer of the network)

```python
with tf.name_scope("loss_fn"):
    
    # Will get a tensor of loss values per sample
    softentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(
                    labels=y, logits = logits)
    loss = tf.reduce_mean(softentropy)
```

Now that we have defined a loss function, we will set an optimizer to minimize it. 

Note the dependencies:
``optimizer`` <- ``loss`` <- `softentropy` <- `logits` <- X 

Which is why when we run the session, we have to feed in a dictionary to populate the placeholders

```python
learning_rate = 0.01

with tf.name_scope("train"):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
```

```python
with tf.name_scope("eval" ):
    correct = tf.nn.in_top_k(logits , y , 1 )
    accuracy = tf.reduce_mean(tf.cast(correct , tf.float32 ))
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

Time to run the graph inside a session

```python
# import pandas
def get_batches(X,y, batchsize):
    n_batches = len(X) // batch_size
    X_batch = np.array_split(X,batchsize)
    y_batch = np.array_split(y,batchsize)
    return zip(X_batch, y_batch)
        
def shuffle_batch(X, y, batch_size):
    # Shuffle the indices
    rnd_idx = np.random.permutation(len(X))
    n_batches = len(X) // batch_size
    
    # Split the index array into the number of batches
    # Get the list of indexs per batch
    # and yield the values when asked
    for batch_idxs in np.array_split(rnd_idx, n_batches):
        X_batch, y_batch = X[batch_idxs], y[batch_idxs]
        yield X_batch, y_batch
```

```python
n_epochs = 10
batch_size = 50

with tf.Session() as sess:
    init.run()
    
    for epoch in range(n_epochs):
        # Run the optimizer once for each batch
        for X_batch,y_batch in shuffle_batch(X_train,y_train,batch_size):
            sess.run(training_op, feed_dict={X: X_batch, y:y_batch})
        
        # Get the accuracy for this epoch using the last batch
        acc_train = accuracy.eval(feed_dict={X: X_batch, y:y_batch})
        
        acc_val = accuracy.eval(feed_dict={X: X_valid, y:y_valid})
        
        print("{} Train accuracy: {:3f}, Val accuracy: {:3f}".format(epoch,acc_train, acc_val))

    save_path = saver.save (sess , "./my_model_final.ckpt")
    
```

If you switch to ReLU and rerun the above network, the accuracies are much lower. So maybe ELU is a good activation function.
Also, decreasing the `learning_rate` to 0.1 makes the NN overfit pretty bad.
We should try adding more layers next.

```python
reset_graph()

n_hidden1 = 300
n_hidden2 = 150
n_outputs = 10

X = tf.placeholder(tf.float32, shape=(None,n_inputs), name = "X")
y = tf.placeholder(tf.int32, shape=(None), name="y")

with tf.name_scope("dnn"):
    hidden1 = tf.layers.dense(X, n_hidden1,name="hidden1",
                              activation=tf.nn.elu)
    hidden2 = tf.layers.dense(hidden1, n_hidden2,name="hidden2",
                              activation=tf.nn.elu)
    hidden3 = tf.layers.dense(hidden2, n_hidden2,name="hidden3",
                               activation=tf.nn.elu)
    logits = tf.layers.dense(hidden3 , n_outputs, name = "outputs" )
    
with tf.name_scope("loss_fn"):
    
    # Will get a tensor of loss values per sample
    softentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(
                    labels=y, logits = logits)
    loss = tf.reduce_mean(softentropy)
    
learning_rate = 0.01

with tf.name_scope("train"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope("eval" ):
    correct = tf.nn.in_top_k(logits , y , 1 )
    accuracy = tf.reduce_mean(tf.cast(correct , tf.float32 ))

init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 40
batch_size = 50

with tf.Session() as sess:
    init.run()
    
    for epoch in range(n_epochs):
        # Run the optimizer once for each batch
        for X_batch,y_batch in shuffle_batch(X_train,y_train,batch_size):
            sess.run(training_op, feed_dict={X: X_batch, y:y_batch})
        
        # Get the accuracy for this epoch using the last batch
        acc_train = accuracy.eval(feed_dict={X: X_batch, y:y_batch})
        
        acc_val = accuracy.eval(feed_dict={X: X_valid, y:y_valid})
        
        print("{} Train accuracy: {:3f}, Val accuracy: {:3f}".format(epoch,acc_train, acc_val))

    save_path = saver.save (sess , "./my_model_final.ckpt")
```
