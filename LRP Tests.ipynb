```python
import tensorflow as tf
import numpy as np
import sys
sys.path.append("interprettensor/")

root_logdir = "tf_logs"

tf.__version__
```

```python
from modules import *
```

```python
# Numpy arrays
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Converting to float and normalizing
X_train = X_train.astype(np.float32).reshape(-1, 28*28) / 255.0
X_test = X_test.astype(np.float32).reshape(-1, 28*28) / 255.0
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)

# Separating out validation set
X_valid, X_train = X_train[:5000], X_train[5000:]
y_valid, y_train = y_train[:5000], y_train[5000:]


n_inputs = 28*28  # MNIST
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

```python
feature_cols = [tf.feature_column.numeric_column("X", shape=[28 * 28])]
dnn_clf = tf.estimator.DNNClassifier(hidden_units=[300,100], n_classes=10,
                                     feature_columns=feature_cols)
```

```python
input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_train}, y=y_train, num_epochs=40, batch_size=50, shuffle=True)

# To save summaries for Tensorboard
with tf.train.MonitoredTrainingSession(summary_dir=root_logdir) as sess:
    dnn_clf.train(input_fn=input_fn)
```

```python
dnn_clf.latest_checkpoint()
```

```python
tf.reset_default_graph()
eval_summary_hook = tf.train.SummarySaverHook(
                                save_steps=100,
                                output_dir= "tf_logs",
                                summary_op=tf.summary.scalar)

# Add it to the evaluation_hook list
with tf.train.MonitoredTrainingSession() as sess:
    dnn_clf.train(input_fn=input_fn, hooks=[eval_summary_hook])
```
