```python
trace0 = go.Scatter(
    x=[1, 2, 3, 4],
    y=[10, 15, 13, 17],
    mode="markers"
)
trace1 = go.Scatter(
    x=[1, 2, 3, 4],
    y=[16, 5, 11, 9],
    mode="lines",
    name="line plot"
)
data = [trace0, trace1]

# py.plotly.plot(data, filename = 'basic-line', auto_open=True)
# py.plotly.iplot(data, filename = 'basic-line')
```
